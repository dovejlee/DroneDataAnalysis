```python
import pandas as pd
import numpy as np
from datetime import datetime
from collections import OrderedDict
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
Anafi_df = pd.read_csv('data/2020_04_16_Anafi_TimeStamp.csv')
Anafi_df = pd.DataFrame.from_dict(OrderedDict(Anafi_df))
```

The "product_gps_latitude" and "product_gps_longitude" values refer to the GPS coordinates of the Anafi drone itself, which are typically used to geotag photos or videos taken by the drone. On the other hand, the "controller_gps_latitude" and "controller_gps_longitude" values refer to the GPS coordinates of the remote controller, which may be used to track the location of the controller or to assist with navigation.

The final latitude and longitude of the drone would depend on what you are trying to do with the data. If you are trying to geotag photos or videos taken by the drone, you would typically use the "product_gps_latitude" and "product_gps_longitude" values as the location of the image or video. If you are trying to track the location of the drone itself, you would use its GPS coordinates. If you are trying to navigate the drone, you may use a combination of the drone's GPS coordinates and the controller's GPS coordinates to determine the drone's location relative to the controller.

```python
product_gps = pd.DataFrame({'latitude': df["product_gps_latitude"], 'longitude' : df["product_gps_longitude"]})
controller_gps = pd.DataFrame({'latitude': df["controller_gps_latitude"], 'longitude' : df["controller_gps_longitude"]})
relative_location = product_gps.subtract(controller_gps)

# Print the result
print(relative_location)
```

```python
fig = plt.figure(figsize=(12,8)) 
axes = fig.add_subplot(1, 1, 1)
# 
# Plot flight data 1 in blue
Controller_Lat = (controller_gps['latitude'])
Controller_Long = (controller_gps['longitude'])
plt.plot(Controller_Long, Controller_Lat, 'y', label='controller_gps')

# Plot flight data 2 in red
Product_Lat = (product_gps['latitude'])
Product_Long = (product_gps['longitude'])
plt.plot(Product_Long, Product_Lat,  'r', label='product_gps')

# Add a title and axis labels
plt.title('Flight Data')
plt.xlabel('Latitude')
plt.ylabel('longitudue')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
