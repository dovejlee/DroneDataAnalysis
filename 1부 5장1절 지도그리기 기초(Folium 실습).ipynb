# 대한민국 행정구역 다운로드 페이지 에서 자료 다운
# QGIS 설치하여 지도 영역 json파일로 만들기
# 지역 법적동 확인- 태안 44825
# mapshaping mapshaper.org
# https://www.youtube.com/watch?v=c0EQWKN5kh4

```python
HanseoUniv=[36.59533952408117, 126.29442554748438]

folium.Map(
    location = HanseoUniv,
    tiles='Stamen Terrain',
    zoom_start=15)  # 15

geo_data = 'TAEAN.geojson'
with open(geo_data, mode='rt', encoding='utf-8') as f:
    geo = json.loads(f.read())
    f.close()
folium.GeoJson(
    geo,
    name='TAEAN_GEO_DATA'
).add_to(m)

folium.Marker(
    location = HanseoUniv,
    popup="한서대",
    icon = folium.Icon(color="blue", icon ="star")
    ).add_to(m)
folium.CircleMarker(
    location = HanseoUniv,
    radius=20,
    color = '#FFF',
    fil_color= "#fffggg",
    popup="한서대태안캠퍼스",
    icon= folium.Icon(color="green", icon= "star")
    ).add_to(m)
m
```

```python
m = folium.Map([37,127], zoom_start=5)

wind_positions=[[32.587385, 133.058046],
                [34.359654, 130.083293],
                [35.570943, 127.200923],
                [36.214572, 124.226170],
                [36.348622, 121.916579],
                [36.645685, 117.981036]]
#2차원 의 행렬 

wind_line = folium.PolyLine(
    wind_positions,
    weight=10,
    color='deepskyblue'
).add_to(m)

plugins.PolyLineTextPath(
    wind_line,
    '> ',
    repeat=True,
    # False/True
    offset=5,
    # change the offset 7, 10, 
    attributes={'fill': 'dodgerblue', 'front_weight': 'bold', 'front-size': '24'}
).add_to(m)

m
```

```python
marine_currents=[[42.235235, 131.951693],
                 [41.391475, 130.442223],
                 [39.775207, 128.887468],
                 [38.758520, 128.691237],
                 [37.571772, 129.385593],
                 [36.256336, 129.778056]]
m= folium.Map()

folium.plugins.AntPath(
    locations=marine_currents,
    reverse='True',
    #True/False
    dash_array=[20,20]
).add_to(m)

m.fit_bounds(m.get_bounds())

m
```

```python
HanseoUniv = [36.59533952408117, 126.29442554748438]
NaMoonJae= [36.5869785, 126.375217]
KkotJiBeach = [36.500569, 126.3373923]
CheongPoDae=[36.63972121453488, 126.30102826869934]

m=folium.Map(location=HanseoUniv,zoom_starts=6)

gs=folium.FeatureGroup(name='Groups')
m.add_child(gs)
g1 = plugins.FeatureGroupSubGroup(gs,'Group1')
m.add_child(g1)
g2 = plugins.FeatureGroupSubGroup(gs,'Group2')
m.add_child(g2)

folium.Marker(NaMoonJae).add_to(g1)
folium.Marker(KkotJiBeach).add_to(g1)
folium.Marker(CheongPoDae).add_to(g2)
#folium.Marker([2,-2]).add_to(g1)
folium.LayerControl(collapsed=False).add_to(m)

m
```
