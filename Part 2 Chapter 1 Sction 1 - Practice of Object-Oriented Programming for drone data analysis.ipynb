```python
import pandas as pd
import numpy as np
import math

# This class represents the calculation process for convenience.
# Other file handling and mathematical operations are performed using pandas and math modules.
# Note: If using Visual Studio Code, ensure the interpreter is set to the Anaconda version of Python to avoid potential pandas errors.

class Math():
    def __init__(self,df,target):
        self.list = df.loc[:,target].to_list()
        self.length_of_list = len(self.list)
        self.sum_list = sum(self.list)

    def average(self):
        self.average = self.sum_list/self.length_of_list
        return self.average
   
    def standard_deviation(self): 
        self.v = []
        for i in range(len(self.list)):
            f = (self.list[i] - self.sum_list/self.length_of_list)**2
            self.v.append(f)
        v = sum(self.v)/self.length_of_list  # Standard_Deviation
        return math.sqrt(v) # Return to Standard_Deviation
             
    
df = pd.read_csv("data/RotaryWing.csv")

length = Math(df=df, target="Length") # instance
width = Math(df=df, target="Width") # instance
weight = Math(df=df, target="Weight") # instance
 
#average compare two method
print("\n\n< Lentgh >")
print("-------Average-------")
print(f"Calculate the average using a Pandas class : {length.average()}")
print(f"Calculate the average using the numpy library : {np.average(length.list)}")

# Standard Deviation
print("-----Standard_Deviation-----")
print(f"Calculate the Standard_Deviation using a Pandas class :  {length.standard_deviation()}")
print(f"Calculate the Standard_Deviation using the numpy library :  {np.std(length.list)}")
print()

#average compare two method
print("\n\n< Width >")
print("-------Average-------")
print(f"Calculate the average using a Pandas class :  {width.average()}")
print(f"Calculate the average using the numpy library : {np.average(width.list)}")
print()

# Standard Deviation
print("-----Standard_Deviation-----")
print(f"Calculate the Standard_Deviation using a Pandas class : {width.standard_deviation()}")
print(f"Calculate the Standard_Deviation using the numpy library : {np.std(width.list)}")

#average compare two method
print("\n\n< Weight >")
print("-------Average-------")
print(f"Calculate the average using a Pandas class :  {weight.average()}")
print(f"Calculate the average using the numpy library : {np.average(weight.list)}")
print()

# Standard Deviation
print("-----Standard_Deviation-----")
print(f"Calculate the average using a Pandas class :  {weight.standard_deviation()}")
print(f"Calculate the Standard_Deviation using the numpy library : {np.std(weight.list)}")
```
