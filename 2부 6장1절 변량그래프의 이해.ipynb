```python
import pandas as pd
import seaborn as sns
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
UAV_df = pd.read_csv("UavSize.csv")
```

```python
UAV_df
```

```python
FixedWing_df = UAV_df[UAV_df["Type"]=="FixedWing"]
MultiCopter_df = UAV_df[UAV_df["Type"]=="MultiCopter"]
```

```python
FixedWing_df
```

```python
MultiCopter_df
```

# 이재용

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,1,1)
axes2 = fig.add_subplot(2,1,2)
```

```python
axes1.plot(FixedWing_df["Weight"], FixedWing_df["Length"], 'o',c ="red")
axes2.plot(MultiCopter_df["Weight"], MultiCopter_df["Length"], 'o', c ='blue' )
fig
```

```python
axes1.set_title("Fixed Wing")
axes2.set_title("Multi Copter")
fig
```

```python
fig.suptitle("UAV Size <Hight-Weight>")
fig
```

```python
fig.tight_layout()
fig
```
