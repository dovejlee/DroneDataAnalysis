```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the airspeed_0.csv file
airspeed_data = pd.read_csv('data/PX4/07_51_52_airspeed_0.csv')

# Calculate the average true airspeed
average_true_airspeed = airspeed_data['true_airspeed_m_s'].mean()

# Print the result
print(f'Average True Airspeed: {average_true_airspeed:.2f} m/s')
```

```python
# Define the rolling window size (e.g., 10 data points)
window_size = 10

# Calculate the rolling average of true airspeed
airspeed_data['rolling_avg_true_airspeed'] = airspeed_data['true_airspeed_m_s'].rolling(window=window_size).mean()

# Plot the original true airspeed and the rolling average
plt.figure(figsize=(12, 6))
plt.plot(airspeed_data['timestamp'], airspeed_data['true_airspeed_m_s'], label='True Airspeed', color='blue')
plt.plot(airspeed_data['timestamp'], airspeed_data['rolling_avg_true_airspeed'], label='Rolling Average Airspeed', color='red', linewidth=2)
plt.xlabel('Timestamp')
plt.ylabel('Airspeed (m/s)')
plt.legend()
plt.show()
```
