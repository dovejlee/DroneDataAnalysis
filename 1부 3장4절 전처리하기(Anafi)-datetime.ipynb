```python
Anafi_df['FlightTime'] = pd.to_datetime(Anafi_df['Date/Time'], format = "%M:%S.%f")
```

```python
AnafiFirst_df = Anafi_df.iloc[:1527,:]
AnafiSecond_df = Anafi_df.iloc[1527:,:]
```

```python
AnafiFirst_df['FlightTimeEdit']=AnafiFirst_df['FlightTime'].apply(lambda x:x.replace(year = 2020, month = 4, day=16, hour = 10))
```

```python
print(AnafiFirst_df['FlightTimeEdit'].dtype)
```

```python
AnafiSecond_df['FlightTimeEdit']=AnafiSecond_df['FlightTime'].apply(lambda x:x.replace(year = 2020, month = 4, day=16, hour = 11))
```

```python
AnafiSecond_df['FlightTimeEdit']
```

```python
AnafiNew_df = pd.concat([AnafiFirst_df, AnafiSecond_df])
```

```python
AnafiNew_df
```

```python
Time_Ser =AnafiNew_df.groupby('FlightTimeEdit', sort = False)['FlightTimeEdit'].min()
```

# Problem description
# DataFrame.groupby() has a parameter that selects whether the result should be sorted by groups.

# However, if the DataFrame has a multi-index and the grouping is only done by one index, the result is sorted regardless of the value of sort.
# Grouping by more than one index works.
# Passing the single index as a list [0] does not fix the problem.

```python
Time_Ser
```

```python
print(Time_Ser.dtype)
```

```python
battery_level =AnafiNew_df.groupby('FlightTimeEdit', 
                                   sort = False)['battery_level'].max() 
```

```python
wifi_signal =AnafiNew_df.groupby('FlightTimeEdit', 
                                 sort = False)['wifi_signal'].min() 
```

```python
# battery level 작업을 위한 전처리
AnafiDateSet=list(zip(Time_Ser, battery_level, lat_Ser, lon_Ser, alt_Ser,
                      wifi_signal, speed_vx, speed_vy, speed_vz, speed ))

df=pd.DataFrame(data=AnafiDateSet, columns=['FlightTimeEdit', 'battery_level', 
                                        'product_gps_latitude','product_gps_longitude', 
                                        'altitude', 'wifi_signal',
                                        'speed_vx', 'speed_vy', 'speed_vz', 'speed'])
```
