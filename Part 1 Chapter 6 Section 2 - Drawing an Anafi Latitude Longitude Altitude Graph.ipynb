Change data frame to Series.. draw except numpy..

```python
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('latitude')
axes.set_xlabel('Time') 
axes.set_ylabel('latitude')

Time = (Anafi_df['FlightTimeEdit'])
Lat = (Anafi_df['product_gps_latitude'])
plt.plot(Time, Lat)
plt.show()
```

```python
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('longitude')
axes.set_xlabel('Time') 
axes.set_ylabel('longitude')

Time = (Anafi_df['FlightTimeEdit'])
Lon = (Anafi_df['product_gps_longitude'])
plt.plot(Time, Lon)
plt.show()
```

MSL altitude in drone data typically refers to the altitude above mean sea level. This means that the altitude is measured from a fixed reference point, which is the average height of the sea at a particular location.

The full name of MSL is "Mean Sea Level". It refers to the average height of the sea surface over a period of time, typically measured over a period of several years. MSL is often used as a reference point for measuring altitude or elevation, particularly for locations near the coast or other bodies of water.

```python
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('altitude')
axes.set_xlabel('Time') 
axes.set_ylabel('altitude')

Time = (Anafi_df['FlightTimeEdit'])
Alt = (Anafi_df['altitude'])
plt.plot(Time, Alt)
plt.show()
```
