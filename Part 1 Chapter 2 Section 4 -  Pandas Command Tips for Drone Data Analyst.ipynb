#### 기체 중량 2kg 이상인 드론을 찾아내어 초경량비행장치자격이 요구되는 지를 찾아 내자
##### Case 1
종류(clsss)  기체 중량                                      </br>
1종          최대이륙중량 25kg 초과 (자체중량 150kg 이하)     </br>
2종          최대이륙중량 7kg 초과 25kg 이하  </br>
3종          최대이륙중량 2kg 초과 7kg 이하</br>
4종          최대이륙중량 250g 초과 2kg 이하</br>
250g         이하 별도 자격증 필요없음.

```python
df['MTOW(g)']
```

```python
df
```

```python
classes = []

for mtow in df['MTOW(g)']:
    if mtow > 25000:
        classes.append('Class 1')
    elif mtow > 7000:
        classes.append('Class 2')
    elif mtow > 2000:
        classes.append('Class 3')
    elif mtow > 250:
        classes.append('Class 4')
    else:
        classes.append('No Required')
        
df['ClassNo'] = classes
```

```python
df
```

#### 기체 중량 2kg 이상인 드론을 찾아내어 초경량비행장치자격이 요구되는 지를 찾아 내자
##### Case 2
종류(clsss)  기체 중량                                      </br>
1종          최대이륙중량 25kg 초과 (자체중량 150kg 이하)     </br>
2종          최대이륙중량 7kg 초과 25kg 이하  </br>

4종          최대이륙중량 250g 초과 2kg 이하</br>
250g         이하 별도 자격증 필요없음.

```python
def make_class(x):
    if x > 25000 :
        return 'Class 1'
    elif x > 7000 :
        return 'Class 2'
    elif x > 2000 :
        return 'Class 3'
    elif x > 250:
        return 'Class 4'
    else:
        return 'No Required'
```

```python
df['Class']=df['MTOW(g)'].apply(make_class)
```

### apply 함수에 파라미터 전달하기 <br/>
키워드 파라미터를 사용하시면, apply가 적용된 함수에 파라미터를 전달할 수 있습니다.

```python
def calculate_year(year, current_year):
    return current_year - int(year)
```

```python
df['oldest'] = df['year'].apply(calculate_year, current_year=2023)
df
```

#### apply 함수에 한 개 이상의 파라미터 전달하기 </br>
키워드 파라미터를 추가해주시면, 원하시는만큼의 파라미터를 함수에 전달 가능합니다.

```python
def get_introduce(passed, prefix, suffix):
    return prefix + str(passed) + suffix
```
