```python
Mavic2_df = pd.read_csv('2018_04_19_Mavic2.csv')
Mavic2_df = pd.DataFrame.from_dict(OrderedDict(Mavic2_df))
```

```python
Mavic2_df
```

```python
# 데이타 프레임을 Series로 바꾸고.. numpy 제외 하고 그리기..
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('Battery Level(Mavic2)')
axes.set_xlabel('Time') 
axes.set_ylabel('BatteryPower(%)')
# ()를 이용하여 데이타 프레임을 시리즈로 변경
Time = (Mavic2_df['Time(seconds)'])
Bat = (Mavic2_df['BatteryPower(%)'])
plt.plot(Time, Bat)
plt.show()
```

```python
# 데이타 프레임을 Series로 바꾸고.. numpy 제외 하고 그리기..
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('Battery Voltage (Mavic2)')
axes.set_xlabel('Time') 
axes.set_ylabel('Battery Voltage')
# 
Time = (Mavic2_df['Time(seconds)'])
BatVolt = (Mavic2_df['BatteryVoltage'])
plt.plot(Time, BatVolt)
plt.show()
```

```python
# 
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('Battery Voltage Deviation (Mavic2)')
axes.set_xlabel('Time') 
axes.set_ylabel('Battery Deviation')
# 
Time = (Mavic2_df['Time(seconds)'])
BatVolt_Devi = (Mavic2_df['BatteryVoltageDeviation'])
plt.plot(Time, BatVolt_Devi)
plt.show()
```
