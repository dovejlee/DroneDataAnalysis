```python
import pandas as pd
import numpy as np
from datetime import datetime
from collections import OrderedDict
```

```python
Anafi_df = pd.read_csv('data/2020_04_16_Anafi.csv')
Anafi_df = pd.DataFrame.from_dict(OrderedDict(Anafi_df))
```

```python
Anafi_df['FlightTime'] = pd.to_datetime(Anafi_df['Date/Time'], format = "%M:%S.%f")
```

```python
print(Anafi_df['FlightTime'].dtype)
```

```python
# Define the target datetime value
target_datetime = pd.Timestamp("1900-01-01 00:00:00")
```

```python
# Create a new column to store the comparison results
Anafi_df["is_equal"] = np.zeros(len(Anafi_df), dtype=bool)

# Loop through each row in the DataFrame
flag = 0
for i in range(len(Anafi_df)):
    # Compare the "full_datetime" value to the target datetime value
    if Anafi_df.iloc[i]["FlightTime"] == target_datetime :
        index = i;
```

```python
print(index)
```

```python
AnafiFirst_df = Anafi_df.iloc[:index,:]
AnafiSecond_df = Anafi_df.iloc[index:,:]
```

```python
AnafiFirst_df['FlightTimeEdit']=AnafiFirst_df['FlightTime'].apply(lambda x:x.replace(year = 2020, month = 4, day=16, hour = 12))
```

```python
AnafiSecond_df['FlightTimeEdit']=AnafiSecond_df['FlightTime'].apply(lambda x:x.replace(year = 2020, month = 4, day=16, hour = 13))
```

```python
AnafiNew_df = pd.concat([AnafiFirst_df, AnafiSecond_df])
```
