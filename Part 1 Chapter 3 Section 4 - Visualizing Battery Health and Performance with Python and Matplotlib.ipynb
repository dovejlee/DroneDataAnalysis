```python
Mavic2_df = pd.read_csv('data/2018_04_19_Mavic2.csv')
Mavic2_df = pd.DataFrame.from_dict(OrderedDict(Mavic2_df))
```

```python
Mavic2_df
```

```python
# Change the Dataframe to Series and drawing except numpy
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('Battery Level(Mavic2)')
axes.set_xlabel('Time') 
axes.set_ylabel('BatteryPower(%)')
#
Time = (Mavic2_df['Time(seconds)'])
Bat = (Mavic2_df['BatteryPower(%)'])
plt.plot(Time, Bat)
plt.show()
```

```python
# Change the Dataframe to Series and drawing except numpy
fig = plt.figure() 
axes = fig.add_subplot(1, 1, 1)
axes.set_title('Battery Voltage (Mavic2)')
axes.set_xlabel('Time') 
axes.set_ylabel('Battery Voltage')
# 
Time = (Mavic2_df['Time(seconds)'])
BatVolt = (Mavic2_df['BatteryVoltage'])
plt.plot(Time, BatVolt)
plt.show()
```
