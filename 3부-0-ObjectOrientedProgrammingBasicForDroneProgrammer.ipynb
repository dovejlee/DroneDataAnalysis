### Method 3

```python
import pandas as pd
import numpy as np
import math

# 편의를 위해 연산 과정만 클래스로 만들었고 
# 그외 파일핸들링, 일부 필요한 수학식은 math와 pandas를 이용했습니다
# visual studio code를 사용하시는 분들은 pandas 오류가 발생할 수 있기에 인터프리터 설정을 아나콘다버전 파이썬으로 변경해 주시면 됩니다

class Math():
    def __init__(self,df,target):
        self.list = df.loc[:,target].to_list()
        self.length_of_list = len(self.list)
        self.sum_list = sum(self.list)

    def average(self): #평균
        self.average = self.sum_list/self.length_of_list
        return self.average
   
    def standard_deviation(self): #표준편차
        self.v = []
        for i in range(len(self.list)):
            f = (self.list[i] - self.sum_list/self.length_of_list)**2
            self.v.append(f)
        v = sum(self.v)/self.length_of_list #분산
        return math.sqrt(v) #표준편차 반환
             
    
df = pd.read_csv("RotaryWing.csv")

length = Math(df=df, target="Length") # 인스턴스
width = Math(df=df, target="Width") # 인스턴스
weight = Math(df=df, target="Weight") # 인스턴스
 
#평균 동일
print("\n\n< Lentgh >")
print("-------평균-------")
print(f"클래스로 구한 평균 : {length.average()}")
print(f"넘파이를 이용해 구한 평균 : {np.average(length.list)}")

# 표준편차 동일
print("-----표준편차-----")
print(f"클래스로 구한 표준편차 : {length.standard_deviation()}")
print(f"넘파이를 이융해 구한 표준편차 : {np.std(length.list)}")


print("\n\n< Width >")
print("-------평균-------")
print(f"클래스로 구한 평균 : {width.average()}")
print(f"넘파이로 구한 평균 : {np.average(width.list)}")
print()

# 표준편차 동일
print("-----표준편차-----")
print(f"클래스로 구한 표준편차 : {width.standard_deviation()}")
print(f"넘파이로 구한 표준편차 : {np.std(width.list)}")


print("\n\n< Weight >")
print("-------평균-------")
print(f"클래스로 구한 평균 : {weight.average()}")
print(f"넘파이로 구한 평균 : {np.average(weight.list)}")
print()

# 표준편차 동일
print("-----표준편차-----")
print(f"클래스로 구한 표준편차 : {weight.standard_deviation()}")
print(f"넘파이로 구한 표준편차 : {np.std(weight.list)}")
```
