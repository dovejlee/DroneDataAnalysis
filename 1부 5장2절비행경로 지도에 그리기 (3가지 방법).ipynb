```python
center = [df.lat.mean(), df.lon.mean()]  #위도 경도 평균값을 중앙좌표로
# min(), max() 가장 동쪽과 서쪽, 남쪽과 북쪽을 중앙에 둘수 있다.
m = f.Map(location=center, zoom_start=16)

for i in df.index[:len(df)]:
    lat = df.loc[i, 'lat']
    long = df.loc[i,'lon']
    f.Circle([lat,long],
        color = 'red',
        radius = 0.1,
        fill=True
    ).add_to(m)
m
```
